```python
import pandas as pd

data= pd.read_csv("chemicals-in-cosmetics-.csv")
data.head(20)
data.shape[0]
data.head()
```

```python
#cleaning by dropping columns without data and dropping unwanted columns
data=data.dropna()
data=data.drop(columns=['MostRecentDateReported','DiscontinuedDate','ChemicalCreatedAt','ChemicalUpdatedAt','ChemicalDateRemoved'],axis=1)
data.head()
```

```python
#changing names
data=data.rename(columns={'CDPHId': 'Cali Health'})
data.head()
```

```python
data1=pd.read_csv("chemicals-in-cosmetics-5.csv")
data1.head()
```

```python
#merging data
df=data.merge(right=data1, left_on='PrimaryCategory', right_on='Primary Category')
df.head(20)
```

```python
#groupby
sub_group = df.groupby('ChemicalName').count

sub_group()
```

```python
#groupby aggregate
df.groupby('CompanyName').agg({'ChemicalCount': 'sum'})
```
